RESIZING

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img=mpimg.imread('./data/Fish/Fish0.jpg')
print( img.shape)
x =np.resize(img,(100,100,3))
print(x.shape)
imgplot = plt.imshow(x.astype(np.uint8))
```

```python
from scipy.misc import imresize
intep_test = ["nearest", "lanczos", "bilinear", "bicubic", "cubic"]

for interpolation in intep_test:
    x = imresize(img, (32, 32) , interp = interpolation)
    fig, axes = plt.subplots()
    plt.axis('off')
    axes.set_title(interpolation)
    axes.imshow(x.astype(np.uint8))
```
